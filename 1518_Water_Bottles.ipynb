```python
class Solution:
    def numWaterBottles(self, numBottles: int, numExchange: int) -> int:
        d = numBottles
        ans = d
        while (d>=numExchange):
            r = d//numExchange
            ans = ans+r
            l = d-(numExchange*r)
            
            d = d//numExchange
            d = d+l
            print(ans)
        return ans
```
