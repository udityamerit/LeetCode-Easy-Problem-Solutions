#### Solution 1

```python
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        num = nums.copy()
        if target not in nums:
            num.append(target)
            num.sort()   
        else:
            for i in range(len(num)):
                if num[i]==target:
                    return i

        for i in range(len(num)):
            if num[i]==target:
                return i
           
```
