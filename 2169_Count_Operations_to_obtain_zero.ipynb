#### Solution 1

```python
class Solution:
    def countOperations(self, num1: int, num2: int) -> int:
        c = 0
        while (num1 != 0 and num2 !=0):
            if (num1 >= num2):
                num1 = num1-num2
                c +=1
            else:
                num2 = num2-num1
                c +=1
        return c
```

#### Solution 2

```python
class Solution:
    def countOperations(self, num1: int, num2: int) -> int:
        c = 0
        while True:
            if num1==0 or num2==0:
                return 0
                
            if (num1 >= num2):
                num1 = num1-num2
                c +=1
            else:
                num2 = num2-num1
                c +=1
            if (num1 == 0 or num2 ==0):
                return c
```
