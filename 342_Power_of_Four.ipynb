#### Solution 1

```python
class Solution:
    def isPowerOfFour(self, n: int) -> bool:
        if n<=0:
            return False
        value = log10(n)/log10(4)
        print(value)
        return value == int(value)
```

#### Solution 2

```python
class Solution:
    def isPowerOfFour(self, n: int) -> bool:
        if n<1:
            return False
        
        while(n%4==0):
            n = n/4
            
        return n==1
```
