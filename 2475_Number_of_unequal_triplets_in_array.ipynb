#### Solution 1

```python
class Solution:
    def unequalTriplets(self, nums: List[int]) -> int:
        l = []
        for i in range(len(nums)):
            for j in range(len(nums)):
                for k in range(len(nums)):
                    if i <j < k and  nums[j]!=nums[i] and nums[k]!=nums[j] and nums[i]!=nums[k]:
                        c = [nums[i],nums[j],nums[k]]
                        # c.sort()
                        # print(c)
                        l.append(c)
                    else:
                        continue
        # print(l)
        return len(l)
```

#### Solution 2

```python
class Solution:
    def unequalTriplets(self, nums: List[int]) -> int:
        
        ans = 0
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                for k in range(j+1, len(nums)):
                    if  nums[j]!=nums[i] and nums[k]!=nums[j] and nums[i]!=nums[k]:
                        ans = ans+1
        return ans
```
