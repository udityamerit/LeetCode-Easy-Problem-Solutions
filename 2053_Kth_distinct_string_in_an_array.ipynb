#### Solution 1

```python
from collections import Counter
class Solution:
    def kthDistinct(self, arr: List[str], k: int) -> str:
        arr = Counter(arr)
        l = []
        for key, value in arr.items():
            if value==1:
                l.append(key)
        print(l)
        if len(l)>=k:

            return l[k-1]
        else:
            return ""
```
