#### Solution 1

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        
        for i in range(len(nums)-1):
            for j in range(i+1,len(nums)):
                if nums[i]>nums[j]:
                    nums[i], nums[j] = nums[j], nums[i]
        
        return nums
        
```

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        
        for i in range(len(nums)-1):
            for j in range(i+1,len(nums)):
                if nums[i]>nums[j]:
                    temp = nums[i]
                    nums[i] = nums[j]
                    nums[j]=temp
                    
        
```
