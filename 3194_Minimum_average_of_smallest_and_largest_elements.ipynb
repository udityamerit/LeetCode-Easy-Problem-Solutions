#### Solution 1

```python
class Solution:
    def minimumAverage(self, nums: List[int]) -> float:
        k = len(nums) // 2
        nums.sort()
        output = []
        for i in range (0,k):
            avg = (nums[0] + nums[-1]) / 2
            output.append(avg)
            del nums[0]
            del nums[-1]
        return min(output)
```
