#### Solution 1

```python
class Solution:
    def canMakeArithmeticProgression(self, arr: List[int]) -> bool:
        arr.sort()
        comm_diff = abs(arr[0]-arr[1])
        for i in range(1, len(arr)-1):
            if(abs(arr[i+1]-arr[i]) != comm_diff):
                return False
        
        return True
```
