```python
class Solution:
    def alternateDigitSum(self, n: int) -> int:
        l = [int(i) for i in str(n)]
        print(l)
        k = []
        for i in range(0,len(l)):
            if i%2==0:
                k.append(l[i])
            else:
                k.append(-l[i])
        print(k)
        return sum(k)
```
