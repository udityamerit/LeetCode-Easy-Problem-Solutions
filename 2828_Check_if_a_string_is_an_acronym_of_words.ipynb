```python
class Solution:
    def isAcronym(self, words: List[str], s: str) -> bool:
        l = []
        if len(words) != len(s):
            return (False)
        else:
            for i in range(len(words)):
                if words[i][0] == s[i]:
                    # print( words[i][0])
                    l.append(True)
        return (len(l)==len(s))

        
```
