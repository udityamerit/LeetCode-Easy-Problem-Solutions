```python
class Solution:
    def findMissingAndRepeatedValues(self, grid: List[List[int]]) -> List[int]:

        l = []
        n = len(grid)
        
        for i in grid:
            l+=i
        print(l)
        ans = []
        st = set(l)
        # print(st)
        l1 = list(st)
        print(l1)
        for i in range(len(l)):
        #     print(i)
            if l.count(l[i])>1:
                ans.append(l[i])
                print(l[i])
        print(ans)
        ans = list(set(ans))
        print(ans)

        for i in range(1,(n**2)+1):
            if i not in l:
                ans.append(i)
        return (ans)
```
