```python
class Solution:
    def judgeSquareSum(self, c: int) -> bool:
        # l = [i for i in range(1, c+1//2)]
        # print(l)
        # for i in l:
        #     for j in l:
        #         print(i,j)
        #         if i**2+ j**2==c:
        #             return True
        
        # return False
        a = 0
        while a * a <= c:
            b = math.isqrt(c - a * a)
            if a * a + b * b == c:
                return True
            a += 1
        return False
        
```
