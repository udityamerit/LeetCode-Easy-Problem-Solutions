#### Solution 1

```python
class Solution:
    def findLucky(self, arr: List[int]) -> int:
        l = []
        for i in range(len(arr)):
            if arr.count(arr[i])==arr[i]:
                l.append(arr[i])
        print(l)
        l.sort(reverse = True)
        if len(l)!=0:
            return l[0]
        else:
            return -1
```

#### Solution 2

```python
from collections import Counter
class Solution:
    def findLucky(self, arr: List[int]) -> int:
        
        arr = Counter(arr)
        ans = -1
        for key, value in arr.items():
            if key==value:
                ans = max(ans,key)
        return ans
```
