```python
class Solution:
    def findComplement(self, num: int) -> int:
        l = bin(num)
        l = (l[2::])
        l = [i for i in str(l)]
        print(l)
        k = []
        for i in range(len(l)):
            if l[i]=='0':
                k.append('1')
            if l[i]=='1':
                k.append('0')
        print(k)
        k = str("".join(k))
        
        decimal = int(k, 2)
        return decimal
```
