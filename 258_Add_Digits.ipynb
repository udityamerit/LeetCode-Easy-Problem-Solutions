#### Solution 1


```python
class Solution:
    def addDigits(self, num: int) -> int:
	
	while num > 9:
		sum = 0
		while num:
			sum += num%10
			num = num//10
			
		num = sum

	return num             
```

#### Solution 2

```python
class Solution:
    def addDigits(self, num: int) -> int:
        while num>9:
            num = [int(i) for i in str(num)]
            num = sum(num)
            num = num
        return num
```
