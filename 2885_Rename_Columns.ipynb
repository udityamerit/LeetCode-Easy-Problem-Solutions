```python
import pandas as pd

def renameColumns(students: pd.DataFrame) -> pd.DataFrame:
    st = pd.DataFrame(students)
    st.rename(columns = {'id':'student_id'}, inplace = True)
    st.rename(columns = {'first':'first_name'}, inplace = True)
    st.rename(columns = {'last':'last_name'}, inplace = True)
    st.rename(columns = {'age':'age_in_years'}, inplace = True)

    return st
```
