#### Solution 1

```python
class Solution:
    def smallestEqual(self, nums: List[int]) -> int:
        
        ans =[]
        for i in range(len(nums)):
            if i%10==nums[i]:
                ans.append(i)
            else:
                continue
        if len(ans)>0:
            return min(ans)
        else:
            return -1
    
```

#### Solution 2


```python
class Solution:
    def smallestEqual(self, nums: List[int]) -> int:
    
        for i in range(len(nums)):
            if i%10==nums[i]:
                return i
          
        return -1
```
