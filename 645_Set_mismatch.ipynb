#### Solution 1

```python
from collections import Counter
class Solution:
    def findErrorNums(self, nums: List[int]) -> List[int]:
        ans = []
        num1 = []
        for i in range(1,len(nums)+1):
            num1.append(i)
          
        num = Counter(nums)
        for key ,value in num.items():
            if num.get(key)>=2:
                ans.append(key)
        
        for i in num1:
            if i not in nums:
                ans.append(i)
        return ans
```
