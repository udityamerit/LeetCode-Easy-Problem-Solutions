```python
class Solution:
    def makeFancyString(self, s: str) -> str:
        ans = s[0]
        print(ans)
        cnt = 1
        for i in range(1, len(s)):
            if s[i] == ans[-1]:
                cnt += 1
                print(ans[-1])
                if cnt < 3:
                    ans += s[i]
            else:
                cnt = 1
                ans += s[i]
        return ans
```
