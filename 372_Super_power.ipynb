#### Solution 1

```python
class Solution:
    def superPow(self, a: int, b: List[int]) -> int:
        c = int("".join([str(x) for x in b]))
        return pow(a,c,1337)
```

#### Solution 2

```python
class Solution:
    def superPow(self, a: int, b: List[int]) -> int:
        c = ""
        for i in b:
            c = c+str(i)
        c = int(c)
        return pow(a,c,1337)
```
