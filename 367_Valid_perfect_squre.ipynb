#### Solution 2 in C++

```python
#include <cmath>

class Solution {
public:
    bool isPerfectSquare(int num) {
        int sqrt_num = static_cast<int>(sqrt(num)); 
        return sqrt_num * sqrt_num == num;  
    }
};
```
