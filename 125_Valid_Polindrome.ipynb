```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = [x for x in s]
        m = []
        for i in range(len(s)):
            if s[i].isalnum()==True:
                m.append(s[i])
        
        m = "".join(m)
        m = m.lower()

        print(m)
        n = m[::-1]
        n = n.lower()
        print(n)
        return m==n
```
