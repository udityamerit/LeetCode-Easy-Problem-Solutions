```python
class Solution:
    def maximumCount(self, nums: List[int]) -> int:
        c = 0
        d = 0
        l = []
        for i in range(len(nums)):
            if nums[i]>0:
                c = c+1
            if nums[i]<0:
                d = d+1
        # l.append(c)
        # l.append(d)
        # return (max(l))
        return max(c,d)
```
