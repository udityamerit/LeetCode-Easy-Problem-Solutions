```python
class Solution:
    def arrayRankTransform(self, arr: List[int]) -> List[int]:
        num = {}
        sortlist = sorted(list(set(arr)))
        for i in range(len(sortlist)):
            num[sortlist[i]] = i+1
            
        for i in range(len(arr)):
            arr[i]= num[arr[i]]
            print(arr[i])

        return arr

```
