#### Solution 1

```python
class Solution:
    def dominantIndex(self, nums: List[int]) -> int:
        x = max(nums)
        l = []
        for i in range(len(nums)):
            if x == nums[i]:
                continue
            else:
                if x>=2*nums[i]:
                    l.append(nums[i])
        print(l)
        if len(l) == len(nums)-1:
            return nums.index(x)
        else:
            return -1
```
