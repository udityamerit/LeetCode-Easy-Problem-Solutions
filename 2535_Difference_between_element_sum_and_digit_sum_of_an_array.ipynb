#### Solution 1

```python
class Solution:
    def differenceOfSum(self, nums: List[int]) -> int:
        l = []
        
        def add_fun(x):
            s = ''
            xz = str(x)
            for i in xz:
                l.append(int(i))

        for i in nums:
            add_fun(i)

        x = sum(nums)
        y = sum(l)
        return abs(x-y)
         
```
