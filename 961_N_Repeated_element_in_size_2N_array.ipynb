#### Solution 1

```python
from collections import Counter
class Solution:
    def repeatedNTimes(self, nums: List[int]) -> int:
        nums = Counter(nums)
        for key, value in nums.items():
            if value>1:
                return key
```

#### Solution 2

```python
from collections import Counter
class Solution:
    def repeatedNTimes(self, nums: List[int]) -> int:
        
        for i in range(len(nums)):
            if nums.count(nums[i])>1:
                return nums[i]
```
