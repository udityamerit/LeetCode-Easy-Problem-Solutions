#### Solution 1 in Python

```python
class Solution:
    def prefixCount(self, words: List[str], pref: str) -> int:

        s= [i for i in pref]
        d = 0

        for i in range(len(words)):
            a = [j for j in words[i]]

            if len(a)>=len(s):
                c = []
                for k in range(len(s)):
                    c.append(a[k])
                    m = "".join(c)
               
                if m==pref:
                    d = d+1
        return d
```

#### Solution 2 in Python

```python
class Solution:
    def prefixCount(self, words: List[str], pref: str) -> int:
        c = 0
        x = len(pref)
        for i in words:
            if i[:x]==pref:
                c = c+1
        return c
```
