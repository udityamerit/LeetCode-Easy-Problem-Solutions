#### Solution 1


```python
class Solution:
    def isPowerOfThree(self, n: int) -> bool:
        if n<=0:
            return False
        value = log10(n)/log10(3)
        print(value)
        return value == int(value)
```

#### Solution 2

```python
class Solution:
    def isPowerOfThree(self, n: int) -> bool:
        if n<=0:
            return False

        while(n%3==0):
            n = n/3
            
        return n==1
```
