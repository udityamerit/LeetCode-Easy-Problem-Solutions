#### Solution 1

```python
class Solution:
    def minOperations(self, logs: List[str]) -> int:
        c = 0
        for i in logs:
            if i == "./":
                continue
            elif i=="../":
                if c==0:
                    c = 0
                else:
                    c = c-1
            
            else:
                c = c+1
                
        if c<=0:
            c = 0
            return c
        else:
            return c
```
