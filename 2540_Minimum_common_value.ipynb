#### Solution 1

```python
from collections import Counter
class Solution:
    def getCommon(self, nums1: List[int], nums2: List[int]) -> int:
        l = []
        nums1 = Counter(nums1)
        nums2 = Counter(nums2)
        for key,value in nums1.items():
            if key in nums2.keys():
                l.append(key)
        if l == []:
            return(-1)
        else:
            return(min(l))
 
```

#### Solution 2

```python

class Solution:
    def getCommon(self, nums1: List[int], nums2: List[int]) -> int:
        
        list1 = []
        if len(nums1)>= len(nums2):
            for i in nums1:
                if i in nums2:
                    list1.append(i)
            return min(list1)

        elif len(nums1)< len(nums2):
            for i in nums2:
                if i in nums1:
                    list1.append(i)
            return min(list1)

        else:
            return -1
                
```
