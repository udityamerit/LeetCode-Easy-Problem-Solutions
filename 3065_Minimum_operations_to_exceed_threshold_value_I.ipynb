#### Solution 1

```python
class Solution:
    def minOperations(self, nums: List[int], k: int) -> int:
        l = []
        nums.sort()
        for i in nums:
            if i<k:
                l.append(i)
        return len(l)
```

#### Solution 2

```python
class Solution:
    def minOperations(self, nums: List[int], k: int) -> int:
        
        c = 0
        for i in nums:
            if i<k:
                c = c+1
        return c
```
