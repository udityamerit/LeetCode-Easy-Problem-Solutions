```python
class Solution:
    def countEven(self, num: int) -> int:
        l = [str(i) for i in range(2,num+1)]
        k = []
        for i in l:
            s = [int(j) for j in i]
            print(s)
            if sum(s)%2==0:
                k.append(i)
        print(k)
        return len(k)

```
