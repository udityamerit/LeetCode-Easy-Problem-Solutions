```python
import sys
class Solution:
    def diagonalPrime(self, nums: List[List[int]]) -> int:
           
        from typing import List


        def is_prime(x):
            if x < 2:
                return False
            for j in range(2, int(x**0.5) + 1):
                if x % j == 0:
                    return False
            return True

        n = len(nums)
        primes = []
        for i in range(n):
            if is_prime(nums[i][i]):
                primes.append(nums[i][i])
            if is_prime(nums[i][n - i - 1]):
                primes.append(nums[i][n - i - 1])

        return max(primes, default=0)

```
