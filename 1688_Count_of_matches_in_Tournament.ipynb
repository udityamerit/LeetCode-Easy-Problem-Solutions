```python
class Solution:
    def numberOfMatches(self, n: int) -> int:
        match_plays = []
        while n != 1:
            if n % 2==0:
                match_plays.append(n/2)
                n = n/2
            else:
                match_plays.append((n-1)/2)
                n = ((n-1)/2)+1
        print(match_plays)
        return int(sum(match_plays))
```
