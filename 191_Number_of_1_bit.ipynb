#### Solution 1

```python
class Solution:
    def hammingWeight(self, n: int) -> int:
        
        x= bin(n)
        x = x[2::]
        l = [int(i) for i in x]
        return sum(l)
```

#### Solution 2

```python
class Solution:
    def hammingWeight(self, n: int) -> int:
        
        return bin(n).count('1')
```
