```python
class Solution:
    def furthestDistanceFromOrigin(self, moves: str) -> int:
        x = 0
        c = moves.count('R')
        c1 = moves.count("L")
        if c>c1:
            for i in moves:
                if i=='L':
                    x -= 1
                elif i=='R':
                    x += 1
                elif i== '_':
                    x +=1
        elif c1>c:
            for i in moves:
                if i=='L':
                    x -= 1
                elif i=='R':
                    x += 1
                elif i== '_':
                    x -=1
        else:
            for i in moves:
                if i== '_':
                    x +=1
                    
        return abs(x)
```
