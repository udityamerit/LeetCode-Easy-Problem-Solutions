```python
class Solution:
    def arithmeticTriplets(self, nums: List[int], diff: int) -> int:
        l = []
        for i in range(len(nums)):
            for j in range(len(nums)):
                for k in range(len(nums)):
                    if i <j < k and  nums[j]- nums[i]== diff and nums[k]-nums[j]==diff:
                        c = [nums[i],nums[j],nums[k]]
                        c.sort()
                        # print(c)
                        if c not in l:

                            l.append(c)
                        else:
                            continue
                    else:
                        continue
        print(l)
        return len(l)
```
