#### Solution 1

```python
class Solution:
    def averageWaitingTime(self, customers: List[List[int]]) -> float:
       
        t = 0
        total_time = 0
        for arr,ord in customers:
            if t>arr:
                total_time = total_time+(t-arr)
            else:
                t = arr
            total_time = total_time+ord
            t = t+ord
        return total_time/len(customers)
            
```
