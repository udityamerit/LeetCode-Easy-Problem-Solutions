#### Solution 1

```python
class Solution:
    def findIndices(self, nums: List[int], indexDifference: int, valueDifference: int) -> List[int]:
        answer = []
        for i in range(len(nums)):
            for j in range(len(nums)):
                if (abs(i-j)>=indexDifference and abs(nums[i]- nums[j])>= valueDifference):
                    answer.append(i)
                    answer.append(j)
                    break
        if len(answer)==0:
            answer = [-1,-1]
        ans = []
        for i in range(2):
            ans.append(answer[i])
        return (ans)
```

#### Solution 2

```python
class Solution:
    def findIndices(self, nums: List[int], indexDifference: int, valueDifference: int) -> List[int]:
        
        for i in range(len(nums)):
            for j in range(len(nums)):
                if (abs(i-j)>=indexDifference and abs(nums[i]- nums[j])>= valueDifference):
                    return [i,j]
        return [-1,-1]
```
