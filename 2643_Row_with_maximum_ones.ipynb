#### Solution 1

```python
class Solution:
    def rowAndMaximumOnes(self, mat: List[List[int]]) -> List[int]:
        l = []
        m = []
        for score in mat:
            l.append(sum(score))
        for i in range(len(mat)):
            if sum(mat[i])== max(l):
                m.append(i)
                m.append(mat[i].count(1))
                break
        return m
```
