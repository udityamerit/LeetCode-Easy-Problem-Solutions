#### Solution 1


```python
class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        
        number = list(str(int("".join([str(x) for x in digits])) + 1))
        return [int(x) for x in number]
        
```

#### Solution 2

```python
class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        
        digit = [str(x) for x in digits]
        d = ""
        for i in digit:
            d = d+i
        d = str(int(d)+1)
        d = [int(x) for x in d]
        return d
```
