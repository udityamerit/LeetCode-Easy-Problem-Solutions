```python
class Solution:
    def capitalizeTitle(self, title: str) -> str:

        t = title.split()

        for i in range(len(t)):
            t[i]=t[i].lower()
            k=t[i].lower()
            if len(t[i])>2:
                p=chr(ord(k[0])-32)+k[1:]
                t[i]=p
        return " ".join(t)
        
```
