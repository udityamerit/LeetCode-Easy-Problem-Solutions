```python
from collections import Counter
class Solution:
    def firstUniqChar(self, s: str) -> int:
        l = []
        g = s
        print(g)
        s = Counter(s)
        print(s)
        for key, value in s.items():
            if value ==1:
                l.append(key)
        
        print(l)
        for i in range(len(g)):
            if len(l)==0:
                return -1
            if l[0]==g[i]:
                return i
            
                
```
