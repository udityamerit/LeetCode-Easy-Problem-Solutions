#### Solution 1

```python
class Solution:
    def mostFrequentEven(self, nums: List[int]) -> int:
        d=dict()
        for i in range(len(nums)):
            if nums[i]%2==0:
                if nums[i] not in d:
                    d[nums[i]]=1
                else:
                    d[nums[i]]+=1
        c=0
        ans=0
        if d:
            for k,v in d.items():
                if v>c:
                    c=v
                    ans=k
                elif v==c:
                    ans=min(ans,k)
            return ans
        return -1
```
