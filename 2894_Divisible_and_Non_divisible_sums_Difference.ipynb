```python
class Solution:
    def differenceOfSums(self, n: int, m: int) -> int:
        divi = []
        non_divi = []
        for i in range(1,n+1):
            if i%m != 0:
                non_divi.append(i)
            else:
                divi.append(i)
        print(divi)
        print(non_divi)
        return ((sum(non_divi)-(sum(divi))))
```
