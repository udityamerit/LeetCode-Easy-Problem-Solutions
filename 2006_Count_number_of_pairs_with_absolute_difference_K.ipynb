```python
class Solution:
    def countKDifference(self, nums: List[int], k: int) -> int:
        c = 0
        for i in range(len(nums)):
            for j in range(len(nums)):
                if j>i:
                    if abs(nums[i]-nums[j])==k:
                        print(nums[i],nums[j])
                        c = c+1
        return c
```
