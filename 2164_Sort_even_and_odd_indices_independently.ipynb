#### Solution 1 in Python

```python
from typing import List

class Solution:
    def sortEvenOdd(self, nums: List[int]) -> List[int]:
        even_indices = nums[::2]  
        odd_indices = nums[1::2]  
        
        even_indices.sort()
        odd_indices.sort(reverse=True)
        
        result = []
        for i in range(len(nums)):
            if i % 2 == 0:
                result.append(even_indices[i // 2])
            else:
                result.append(odd_indices[i // 2])
                
        return result
```

#### Solution 2 in Python


```python
class Solution:
    def sortEvenOdd(self, nums: List[int]) -> List[int]:
        nums[::2] = sorted(nums[::2])
        nums[1::2] = sorted(nums[1::2], reverse = True)
        return nums
```
