```python
class Solution:
    def thirdMax(self, nums: List[int]) -> int:
        nums = list(set(nums))
        nums.sort()
        print(nums)
        if len(nums)==1:
            return nums[0]
        elif len(nums)==2:
            return max(nums)
        elif len(nums)>=3:
            return nums[-3]
```
