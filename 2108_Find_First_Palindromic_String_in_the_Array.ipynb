```python
class Solution:
    def firstPalindrome(self, words: List[str]) -> str:
        res = []
        y = ""
        def rev(s):
            x = [str(x) for x in s]
            for i in range(int(len(x)/2)):
                x[i],x[len(x)-1-i] = x[len(x)-1-i], x[i]
            x = "".join(x)
            res.append(x)
            # print(x)
        for i in words:
            s = rev(i)
        for i in range(len(res)):
            if words[i]==res[i]:
                y = res[i]
                break
        if len(y)>0:
            return str(y)
        else:
            return y
```
