```python
class Solution:
    def findNonMinOrMax(self, nums: List[int]) -> int:
        m = max(nums)
        n = min(nums)
        print(nums)
        print(m, n)
        if len(nums)<=2:
            return -1
        else:
            nums.remove(m)
            nums.remove(n)
            return nums[0]
```
