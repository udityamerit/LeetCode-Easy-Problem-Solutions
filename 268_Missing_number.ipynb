#### Solution 1


```python
class Solution:
    def missingNumber(self, nums: List[int]) -> int:
        
        for i in range(len(nums)+1):
            if i not in nums:
                return i
```

#### Solution 2

```python
class Solution:
    def missingNumber(self, nums: List[int]) -> int:
                
        s = sum(nums)
        c = []
        for i in range(len(nums)+1):
            c.append(i)
        s1 = sum(c)
        return s1-s
```
