```python
class Solution:
    def luckyNumbers (self, matrix: List[List[int]]) -> List[int]:
        r_min= []
        c_max = []
        c = []
        lucky_num=[]
        a = []
        for i in matrix:
            r_min.append(min(i))
        print(r_min)


        for column_index in range(len(matrix[0])):
                column = [row[column_index] for row in matrix]
                c.append((column))
                
        for i in c:
            c_max.append(max(i))
        print(c_max)

        a = c_max+r_min
        print(a)

        for i in range(len(a)):
            if a.count(a[i])>1:
                lucky_num.append(a[i])
        return (list(set(lucky_num)))
```
