#### Solution 1

```python
class Solution:
    def applyOperations(self, nums: List[int]) -> List[int]:
        for i in range(len(nums)-1):
            if nums[i]==nums[i+1]:
                nums[i] = nums[i]*2
                nums[i+1] = 0
            else:
                continue

        l = [i for i in nums if i!=0]
        k = [i for i in nums if i==0]

        return l+k
```
