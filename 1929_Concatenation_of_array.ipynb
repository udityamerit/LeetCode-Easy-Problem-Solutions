```python
class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:
        
        #method 1
        ans = nums+nums
        return ans
```

#### Solution 2

```python
class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:
      
        # method 2:
        ans= []
        for i in range(len(nums)):
            ans.append(nums[i])
        for i in range(len(nums)):
            ans.append(nums[i])
        return ans
```

#### Solution 3


```python
class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:

        #method 3
        
        ans = []
        for i in range(2):
            for i in nums:
                ans.append(i)
        return ans
```
