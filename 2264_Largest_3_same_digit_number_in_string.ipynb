```python
class Solution:
    def largestGoodInteger(self, num: str) -> str:
        l = []
        if len(num)>3:
            for i in range(len(num)-2):
                if num[i]==num[i+1]==num[i+2]:
                    s = str(num[i])+str(num[i+1])+str(num[i+2])
                    print(s)
                    l.append((s))
        if len(num)==3:
            for i in range(1):
                if num[i]==num[i+1]==num[i+2]:
                        s = str(num[i])+str(num[i+1])+str(num[i+2])
                        print(s)
                        l.append((s))
        print(l)
        l.sort(reverse = True)
        print(l)
        if len(l)!=0:
            return str(l[0])
        else:
            return ""
```
