```python
class Solution:
    def nextGreatestLetter(self, letters: List[str], target: str) -> str:
        l = [ord(i)-96 for i in letters]
        print(l)
        c = ord(target)-96
        m = []
        for i in l:
            if i>c:
                m.append( chr(i+96))
        print(m)
        if len(m)==0:
            return letters[0]
        else:
            return m[0]
        
        
```
