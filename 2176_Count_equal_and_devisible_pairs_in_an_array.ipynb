```python
class Solution:
    def countPairs(self, nums: List[int], k: int) -> int:
        c = 0
        for i in range(len(nums)):
            for j in range(len(nums)):
                if (nums[i]==nums[j] and i*j%k==0 and i<j):
                
                    print(nums[i], nums[j], i, j)
                    c = c+1
        return c
```
