#### Solution



```python
import pandas as pd

def findHeavyAnimals(animals: pd.DataFrame) -> pd.DataFrame:
    animals.drop(['species','age'], axis = 1, inplace = True) 
    ani= animals[animals['weight']>100].sort_values(by = "weight", ascending = False)
    ani.drop(['weight'], axis = 1, inplace = True)
    return ani
```
