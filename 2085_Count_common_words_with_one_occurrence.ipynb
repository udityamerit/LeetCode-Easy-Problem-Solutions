```python
class Solution:
    def countWords(self, words1: List[str], words2: List[str]) -> int:
        c = 0
        l=(len(words1))
        m=(len(words2))
        if l>m:
            for i in words1:
                if words1.count(i)==1 and words2.count(i)==1:
                    print(i)
                    c = c+1
        else:
            for i in words2:
                if words2.count(i)==1 and words1.count(i)==1:
                    print(i)
                    c = c+1
        return (c)
```
