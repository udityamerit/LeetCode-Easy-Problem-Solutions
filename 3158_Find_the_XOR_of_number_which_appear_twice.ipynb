#### Solution 1

```python
from collections import Counter
class Solution:
    def duplicateNumbersXOR(self, nums: List[int]) -> int:
        nums = Counter(nums)
        l = []
        for key, value in nums.items():
            if value==2:
                l.append(key)
        print(l)
        if l==[]:
            return 0
        elif len(l)==1:
            return l[0]
        else:
            
            if len(l)>1:
                x = l[0]
                for i in range(len(l)-1):
                    x = x^l[i+1]
                return x
```
