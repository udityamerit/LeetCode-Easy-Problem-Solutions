#### Solution 1

```python
class Solution:
    def separateDigits(self, nums: List[int]) -> List[int]:
        nums = [str(x) for x in nums]
        e = []
        def convert(x):
            l = []
            for i in x:
                l.append(i)
            l = [int(x) for x in l]
            print(l)
            if len(l)>0:
                return l
            else:
                return 0
        for i in nums:
            d = convert(i)
            e.append(d)
        print(e)
        l2 = []
        for i in range(len(e)):
            for j in range(len(e[i])):
                l2.append(e[i][j])
        return l2
```
