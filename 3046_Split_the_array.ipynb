```python
class Solution:
    def isPossibleToSplit(self, nums: List[int]) -> bool:
        nums.sort()        
        nums1 = [nums[i] for i in range(0,len(nums),2)]
        nums2 = [nums[i] for i in range(1,len(nums),2)]
        
        print(nums1)

        print(nums2)
        for i in nums1:
            if (nums1.count(i)>1):
                return False
        for i in nums2:
            if (nums2.count(i)>1):
                return False
        
        return True
```
