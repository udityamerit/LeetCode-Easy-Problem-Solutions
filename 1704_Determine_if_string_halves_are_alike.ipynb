#### Solution 1

```python
class Solution:
    def halvesAreAlike(self, s: str) -> bool:
        a = ""
        b = ""
        c = 0
        j = 0
        vo = ['a','e','i','o','u','A','E','I','O','U']

        # Storing the half string in a 
        for i in range(len(s)//2):
            a = a+s[i]

        # Storing the other half string in b
        for i in range(len(s)//2,len(s)):
            b = b+s[i]
        
        # count the vowels in a string
        for v in a:
            if v in vo:
                c +=1

        # count the vowels in b string
        for v in b:
            if v in vo:
                j +=1

        # Comparing the vowels present in the string a and string b
        if c==j:
            return (True)
        
        return False
```
