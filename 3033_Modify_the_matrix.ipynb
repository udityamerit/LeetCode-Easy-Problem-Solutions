#### Solution 1

```python
class Solution:
    def modifiedMatrix(self, matrix: List[List[int]]) -> List[List[int]]:
        ans = matrix.copy()
        rows, cols= len(matrix), len(matrix[0])
        for c in range(cols):
            currMax = matrix[0][c]
            for r in range(rows):
                currMax = max(matrix[r][c], currMax)
            for r in range(rows):
                if matrix[r][c] == -1:
                    ans[r][c] = currMax
        return ans

        
```
