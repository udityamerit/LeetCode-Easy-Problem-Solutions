#### Solution 1

```python
class Solution:
    def uncommonFromSentences(self, s1: str, s2: str) -> List[str]:
       
        s1= s1.split(" ")
        s2= s2.split(" ")
        sen = s1+s2
        print(sen)
        counter = Counter(sen)
        
        l = [w for w in sen if counter[w]==1]
        return l
```
