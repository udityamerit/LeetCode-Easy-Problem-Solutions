#### Solution 1

```python
class Solution:
    def countPrimeSetBits(self, left: int, right: int) -> int:
        b = [bin(i) for i in range(left,right+1)]
        # print(b)
        l = []
        for i in b:
            k = i[2::]
            l.append(k)
        # print(l)
        a = []
        for i in l:
            s = [int(j) for j in i]
            a.append(sum(s))
        # print(a)
        g = 0
        for i in a:
            c = 0
            for j in range(1,i):
                if i%j==0:
                    c = c+1
            if c==1:
                g = g+1
        return g
```
