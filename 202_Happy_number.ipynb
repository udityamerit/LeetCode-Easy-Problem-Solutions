#### Solution 1

```python
class Solution:
    def isHappy(self, n: int) -> bool:
        
        if n==1 or n==7:
            return True
        if n>9: 
            n = [(int(i))**2 for i in str(n)]
            n = sum(n)
            # print(n)
            while n>=1:
                n = [(int(i))**2 for i in str(n)]
                n = sum(n)
                # print(n)
                if n==1:
                    return True
                else:
                    continue


        else:
            return False   
```

#### Solution 2

```python
class Solution:
    def isHappy(self, n: int) -> bool:
        
        def get_next(n):
            return sum(int(digit) ** 2 for digit in str(n))
        
        s = set()
        while n != 1 and n not in s:
            s.add(n)
            n = get_next(n)
        
        return n == 1
```
