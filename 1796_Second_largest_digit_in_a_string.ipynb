```python
class Solution:
    def secondHighest(self, s: str) -> int:
        l = [int(i) for i in s if i.isdigit()]
        print(l)
        l =list( set(l))
        l.sort(reverse = True)

        print(l)
        if len(l)>1:
            return l[1]
        else:
            return -1
        
```
