#### Solution 1

```python
class Solution:
    def countDistinctIntegers(self, nums: List[int]) -> int:
        c = []
        def reverse_number(x):
            res = 0
            while(x>0):
                r = x%10
                res = res*10+r
                x = x//10
            c.append(res)
        
        for i in nums:
            reverse_number(i)
        for i in range(len(c)):
            nums.append(c[i])
        
        
        return len(set(nums))
```
