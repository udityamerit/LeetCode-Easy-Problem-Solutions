```python
class Solution:
    def judgeCircle(self, moves: str) -> bool:
        # moves = [x for x in list(moves)]
        # for i in range(len(moves)):
        x = 0
        y = 0
        for mov in moves:
            if mov == "U":
                y = y+1
            elif mov == "D":
                y = y-1
            elif mov == "R":
                x = x+1
            else:
                x = x-1
        return x==0 and y == 0
        
```
