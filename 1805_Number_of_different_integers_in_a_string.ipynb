```python
class Solution:
    def numDifferentIntegers(self, word: str) -> int:
        nums='0123456789'

        for i in range(len(word)):
            if word[i] not in nums:
                word= word.replace(word[i],' ')
        num_list=word.split()
        print(num_list)
        res=[]
        for i in range(len(num_list)):
            res.append(int(num_list[i]))
        
        print(res)

        return len(set(res))   
        
                
```
