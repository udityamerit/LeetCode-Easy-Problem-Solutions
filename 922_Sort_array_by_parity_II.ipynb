```python
class Solution:
    def sortArrayByParityII(self, nums: List[int]) -> List[int]:
        e = [i for i in nums if i%2==0]
        o = [i for i in nums if i%2==1]
        print(e)
        print(o)
        j = 0
        k = 0
        ans = []
        for i in range(len(nums)):
            if i%2==0:
                
                ans.append(e[j])
                j = j+1
            else:
                ans.append(o[k])
                k =k+1
        return ans
```
