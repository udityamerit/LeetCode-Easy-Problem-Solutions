```python
from collections import Counter
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        l = []
        k = []
        arr = Counter(arr)
        for key, value in arr.items():
            l.append(value)
        l.sort()
        print(l)
        print(arr)


        for i in range(len(l)-1):
            if l.count(l[i])>1:
                k.append(False)
            else:
                k.append(True)
        l.sort()
        print(l)

        if False in k:
            return False
        else: 
            return True
        
```
