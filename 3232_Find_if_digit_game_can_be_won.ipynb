#### Solution 1

```python
class Solution:
    def canAliceWin(self, nums: List[int]) -> bool:
        l = [i for i in nums if i>=10]
        print(l)
        k = [i for i in nums if i<10]
        print(k)

        if sum(l)>sum(k):
            return True
        elif sum(k)>sum(l):
            return True
        else:
            return False
```
