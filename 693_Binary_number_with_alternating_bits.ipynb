#### Solution 1

```python
class Solution:
    def hasAlternatingBits(self, n: int) -> bool:
        n = bin(n)
        n = n[2::]
        print(n)
        l = []
        for i in range(len(n)-1):
            if n[i] != n[i+1]:
                l.append(True)
            else:
                l.append(False)

        if False in l:
            return False
        else:
            return True
        
```
