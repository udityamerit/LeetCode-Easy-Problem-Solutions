#### Solution 1

```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        address = [i for i in address]
        add =""
        for i in range(len(address)):
            if address[i]==".":
                address[i]="[.]"
        for i in address:
            add = add+i
        return add
```

#### Solution 2

```python
class Solution:
    def defangIPaddr(self, address: str) -> str:

        return address.replace('.','[.]')
```
