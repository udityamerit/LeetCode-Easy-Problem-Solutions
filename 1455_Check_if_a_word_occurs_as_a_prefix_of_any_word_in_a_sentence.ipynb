#### Solution 1 in Python

```python
class Solution:
    def isPrefixOfWord(self, sentence: str, searchWord: str) -> int:
        l = sentence.split()
        s= [i for i in searchWord]
        d = []

        for i in range(len(l)):
            a = [j for j in l[i]]
            if len(a)>=len(s):
                c = []
                for k in range(len(s)):
                    c.append(a[k])
                    m = "".join(c)

                if m==searchWord:
                    d.append(i)
                    
        if d!=[]:
            return min(d)+1
        else:
            return -1
```

#### Solution 2 in Python

```python
class Solution:
    def isPrefixOfWord(self, sentence: str, searchWord: str) -> int:
        # solution 1
        x = len(searchWord)
        l = sentence.split()
        c = []
        for i in range(len(l)):
            # c =0
            if len(l[i])>=x:
                if l[i][:x]==searchWord:
                    c.append(i)
        
        if c==[]:
            return -1
        else:
            return min(c)+1
```
