```python
class Solution:
    def isArraySpecial(self, nums: List[int]) -> bool:
        if len(nums)==1:
            return True
        else:
            l = []
            for i in range(len(nums)-1):
                if ((nums[i]%2==0 and nums[i+1]%2==0) or(nums[i]%2!=0 and nums[i+1]%2!=0)):
                    # l.append([nums[i],nums[i+1]])
                    return False
            return True
        print(l)
```
