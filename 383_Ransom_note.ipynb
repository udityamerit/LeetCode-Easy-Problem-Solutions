```python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        rn = [x for x in ransomNote]
        mag = [x for x in magazine]
        mag.sort()
        rn.sort()
        l = []
        print(rn)
        print(mag)
        if len(rn)<=len(mag):
            for i in range(len(rn)):
                if rn[i] in mag:
                    l.append(True)
                    mag.remove(rn[i])
        print(l)
        if len(l)==len(rn):
            return (True)
        else:
            return (False)
```
