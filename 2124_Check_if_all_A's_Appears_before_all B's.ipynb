#### Solution 1

```python
class Solution:
    def checkString(self, s: str) -> bool:
        l = ""
        k = ""
        for i in s:
            if i=='a':
                l = l+str(i)

        for i in s:
            if i=='b':
                k = k+str(i)               
        
        m = l+k

        return s==m
```

```python
class Solution:
    def checkString(self, s: str) -> bool:
        
        return list(s)==sorted(s)
```
